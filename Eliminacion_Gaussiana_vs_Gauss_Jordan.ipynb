```python
def graficar_sistema_ecuaciones(coeficientes, constantes):
    if len(coeficientes) != 2 or len(coeficientes[0]) != 2 or len(coeficientes[1]) != 2:
        print("Esta función solo soporta sistemas 2D.")
        return

    x = np.linspace(-10, 10, 400)

    a1, b1 = coeficientes[0]
    a2, b2 = coeficientes[1]
    c1, c2 = constantes

    y1 = (c1 - a1 * x) / b1
    y2 = (c2 - a2 * x) / b2

    plt.figure(figsize=(8, 6))
    plt.plot(x, y1, label=f'{a1}x + {b1}y = {c1}')
    plt.plot(x, y2, label=f'{a2}x + {b2}y = {c2}')

    A = np.array(coeficientes)
    B = np.array(constantes)
    try:
        solucion = np.linalg.solve(A, B)
        plt.scatter(solucion[0], solucion[1], color='red', zorder=5)
        plt.text(solucion[0], solucion[1], f'({solucion[0]:.2f}, {solucion[1]:.2f})', fontsize=12, ha='right')
    except np.linalg.LinAlgError:
        print("El sistema no tiene una solución única.")

    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.title('Solución gráfica del sistema de ecuaciones')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(-10, 10)
    plt.xlim(-10, 10)
    plt.show()
```

```python
coeficientes_1 = [[1, 2], [1, -1]]
constantes_1 = [0, 0]
graficar_sistema_ecuaciones(coeficientes_1, constantes_1)
```

```python
def plot_system(a, b):
    try:
        A = np.array([a[0][:2], a[1][:2]])
        B = np.array([b[0], b[1]])

        sol = np.linalg.solve(A, B)

        x = np.linspace(-10, 10, 400)
        y1 = (-a[0][0]*x + b[0]) / a[0][1]
        y2 = (-a[1][0]*x + b[1]) / a[1][1]
        y3 = (-a[2][0]*x + b[2]) / a[2][1]

        plt.figure(figsize=(8, 8))
        plt.plot(x, y1, label=f'{a[0][0]}x + {a[0][1]}y = {b[0]}')
        plt.plot(x, y2, label=f'{a[1][0]}x + {a[1][1]}y = {b[1]}')
        plt.plot(x, y3, label=f'{a[2][0]}x + {a[2][1]}y = {b[2]}')
        plt.scatter(sol[0], sol[1], color='red', zorder=5, label='Intersección de las dos primeras ecuaciones')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Solución gráfica del sistema de ecuaciones')
        plt.grid(True)
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.show()

        return sol

    except np.linalg.LinAlgError:

        x = np.linspace(-10, 10, 400)
        y1 = (-a[0][0]*x + b[0]) / a[0][1]
        y2 = (-a[1][0]*x + b[1]) / a[1][1]
        y3 = (-a[2][0]*x + b[2]) / a[2][1]

        plt.figure(figsize=(8, 8))
        plt.plot(x, y1, label=f'{a[0][0]}x + {a[0][1]}y = {b[0]}')
        plt.plot(x, y2, label=f'{a[1][0]}x + {a[1][1]}y = {b[1]}')
        plt.plot(x, y3, label=f'{a[2][0]}x + {a[2][1]}y = {b[2]}')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Solución gráfica del sistema de ecuaciones')
        plt.grid(True)
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.show()

        return "Sistema sin solución única (matriz singular)"
```

```python
C = [
    [2, 1],
    [1, 1],
    [1, -3]
]
c = [-1, 2, 5]

sol = plot_system(C, c)
print('Solución:', sol)
```

```python
def grafico(a, b, z_value=0):
    try:
        x = np.linspace(-10, 10, 400)

        y1 = (b[0] - a[0][2] * z_value - a[0][0] * x) / a[0][1]
        y2 = (b[1] - a[1][2] * z_value - a[1][0] * x) / a[1][1]

        plt.figure(figsize=(8, 8))
        plt.plot(x, y1, label=f'{a[0][0]}x + {a[0][1]}y + {a[0][2]}z = {b[0]} (z = {z_value})')
        plt.plot(x, y2, label=f'{a[1][0]}x + {a[1][1]}y + {a[1][2]}z = {b[1]} (z = {z_value})')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Solución gráfica del sistema de ecuaciones en 2D (con z = 0)')
        plt.grid(True)
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.show()

        return "Solución gráfica generada"

    except np.linalg.LinAlgError:
        return "Sistema sin solución única (matriz singular)"
```

```python
def gaussian_elimination_with_back_substitution(A, b):
    n = len(b)
    A = A.astype(float)
    b = b.astype(float)
    for i in range(n):
        for j in range(i+1, n):
            factor = A[j,i] / A[i,i]
            for k in range(i, n):
                A[j,k] -= factor * A[i,k]
            b[j] -= factor * b[i]
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = b[i]
        for j in range(i+1, n):
            x[i] -= A[i,j] * x[j]
        x[i] /= A[i,i]
    return np.round(x, 2)
```

```python
def gaussian_elimination(A, b):
    n = len(b)
    A = A.astype(float)
    b = b.astype(float)
    for i in range(n):
        max_row = np.argmax(np.abs(A[i:, i])) + i
        if i != max_row:
            A[[i, max_row]] = A[[max_row, i]]
            b[[i, max_row]] = b[[max_row, i]]
        for j in range(i+1, n):
            factor = A[j, i] / A[i, i]
            A[j, i:] -= factor * A[i, i:]
            b[j] -= factor * b[i]
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = b[i] - np.dot(A[i, i+1:], x[i+1:])
        x[i] /= A[i, i]
    return x
```

```python
A_a = np.array([[1, -1, 3], [3, -3, 1], [1, 1, 0]])
b_a = np.array([2, -1, 3])
sol_a = gaussian_elimination(A_a, b_a)

A_b = np.array([[2, -1.5, 3], [-1, 0, 2], [4, -4.5, 5]])
b_b = np.array([1, 3, 1])
sol_b = gaussian_elimination(A_b, b_b)

A_c = np.array([[2, 0, 0, 0], [1, 1.5, 0, 0], [-3, 0, 0.5, 0], [2, -2, 0, 1]])
b_c = np.array([3, 4.5, -6.6, 0.8])
sol_c = gaussian_elimination(A_c, b_c)

A_d = np.array([[1, 1, 0, 1], [2, 1, -1, 1], [4, -1, -2, 2], [3, -1, -1, 2]])
b_d = np.array([2, 1, 0, -3])
sol_d = gaussian_elimination(A_d, b_d)

print("Solución para el sistema a:", sol_a)
print("Solución para el sistema b:", sol_b)
print("Solución para el sistema c:", sol_c)
print("Solución para el sistema d:", sol_d)
```

```python
import numpy as np

def solve_system(alpha):
    
    A = np.array([[1, -1, alpha],
                  [-1, 2, -alpha],
                  [alpha, 1, 1]])
    b = np.array([-2, 3, 2])
    
    
    det_A = np.linalg.det(A)
    
    if det_A != 0:
        x = np.linalg.solve(A, b)
        return f"α = {alpha}: Solución única, x = {x}"
    else:
        rank_A = np.linalg.matrix_rank(A)
        augmented_matrix = np.column_stack((A, b))
        rank_augmented = np.linalg.matrix_rank(augmented_matrix)
        
        if rank_A == rank_augmented:
            return f"α = {alpha}: Infinitas soluciones"
        else:
            return f"α = {alpha}: No tiene solución"

alphas = [0, 1, -1, 2, -2]

results = [solve_system(alpha) for alpha in alphas]
for result in results:
    print(result)
```

```python
def gauss_jordan(A, b, dtype=np.float32, round_digits=None):
    """
    Resuelve el sistema de ecuaciones lineales Ax = b usando el método de eliminación Gauss-Jordan.
    Si se especifica dtype, usa ese tipo numérico para el cálculo.
    Si se especifica round_digits, redondea los resultados intermedios al número de dígitos dado.
    """
    A = A.astype(dtype)
    b = b.astype(dtype)

    if round_digits:
        A = np.round(A, round_digits)
        b = np.round(b, round_digits)

    n = len(b)

    # Eliminación hacia adelante
    for k in range(n):
        # Asegurar que el pivote es 1
        b[k] /= A[k, k]
        A[k, :] /= A[k, k]

        # Eliminar todas las otras entradas en la columna k
        for i in range(n):
            if i != k:
                factor = A[i, k]
                A[i, :] -= factor * A[k, :]
                b[i] -= factor * b[k]
                if round_digits:
                    A[i, :] = np.round(A[i, :], round_digits)
                    b[i] = round(b[i], round_digits)

    return np.round(b, round_digits)

# Ejercicio 4a con precisión de 32 bits
A4a = np.array([[0.25, 0.2, 0.167], [0.333, 0.25, 0.2], [0.5, 1, 2]], dtype=np.float32)
b4a = np.array([9, 8, 8], dtype=np.float32)

sol_4a = gauss_jordan(A4a, b4a, dtype=np.float32, round_digits=3)
print("Solución 4a:", sol_4a)

A4b = np.array([[3.333, 15920, -10.333], [2.222, 16.71, 9.612], [1.5611, 5.1791, 1.6852]], dtype=np.float32)
b4b = np.array([15913, 28.544, 8.4254], dtype=np.float32)

sol_4b = gauss_jordan(A4b, b4b, dtype=np.float32, round_digits=3)
print("Solución 4b:", sol_4b)

A4c = np.array([[1, 1/2, 1/3, 1/4], [1/2, 1/3, 1/4, 1/5], [1/3, 1/4, 1/5, 1/6], [1/4, 1/5, 1/6, 1/7]], dtype=np.float32)
b4c = np.array([1/6, 1/7, 1/8, 1/9], dtype=np.float32)

sol_4c = gauss_jordan(A4c, b4c, dtype=np.float32, round_digits=3)
print("Solución 4c:", sol_4c)

A4d = np.array([[2, 1, -1, 1, -3], [1, 0, 2, -1, 1], [0, -2, -1, 1, -1], [3, 1, -4, 0, 5], [1, -1, -1, -1, 1]], dtype=np.float32)
b4d = np.array([7, 2, -5, 6, -3], dtype=np.float32)

sol_4d = gauss_jordan(A4d, b4d, dtype=np.float32, round_digits=3)
print("Solución 4d:", sol_4d)
```
