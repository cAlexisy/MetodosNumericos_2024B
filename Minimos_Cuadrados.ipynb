```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

def ajuste_potencia(x, y):
    """
    Ajusta un modelo de potencia de la forma y = b * x^a a los datos proporcionados y calcula el error cuadrático medio.

    Parámetros:
    x (array): Array de valores x.
    y (array): Array de valores y.

    Devuelve:
    None
    """
    # Transformación de los datos para ajuste lineal
    x_log = np.log(x)
    y_log = np.log(y)

    # Ajuste del modelo lineal a los datos transformados
    coeffs = np.polyfit(x_log, y_log, 1)
    a = coeffs[0]
    log_b = coeffs[1]
    b = np.exp(log_b)

    y_fit = b * x**a

    mse = mean_squared_error(y, y_fit)

    print(f"Coeficientes del modelo de potencia: b = {b}, a = {a}")
    print(f"Error cuadrático medio: {mse}")

    plt.scatter(x, y, label='Datos originales')
    plt.plot(x, y_fit, label=f'Ajuste de potencia')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Ajuste de potencia por mínimos cuadrados')
    plt.legend()
    plt.show()

# Datos
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])
```

```python
import matplotlib.pyplot as plt

def graficar_recta_regresion(valores_x, valores_y, etiqueta_x, etiqueta_y, titulo):
    """
    Función para graficar una línea de regresión por mínimos cuadrados.

    valores_x: Lista de valores en el eje x.
    valores_y: Lista de valores en el eje y.
    etiqueta_x: Etiqueta para el eje x.
    etiqueta_y: Etiqueta para el eje y.
    titulo: Título del gráfico.
    """
    # Crear el gráfico de dispersión
    plt.scatter(valores_x, valores_y, color='blue')
    plt.title(titulo)
    plt.xlabel(etiqueta_x)
    plt.ylabel(etiqueta_y)
    plt.grid(True)

    # Cálculo de la línea de regresión por mínimos cuadrados
    n = len(valores_x)
    suma_x = sum(valores_x)
    suma_y = sum(valores_y)
    suma_x_cuadrado = sum(x**2 for x in valores_x)
    suma_xy = sum(x*y for x, y in zip(valores_x, valores_y))

    pendiente = (n * suma_xy - suma_x * suma_y) / (n * suma_x_cuadrado - suma_x**2)
    intercepto = (suma_y - pendiente * suma_x) / n

    def recta_regresion(x):
        return pendiente * x + intercepto

    valores_regresion = [recta_regresion(x) for x in valores_x]

    plt.plot(valores_x, valores_regresion, color='green', linewidth=2)

    plt.show()

    print(f"Ecuación de la recta: y = {pendiente:.3f}x + {intercepto:.3f}")
```
